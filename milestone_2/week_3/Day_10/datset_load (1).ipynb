```python
import pandas as pd

chunks = pd.read_csv("/content/accidents.csv", chunksize=200000)
us_accident_df = pd.concat(chunks)

```

```python
us_accident_df.drop(['End_Lat','End_Lng'],axis=1,inplace=True)
```

```python
us_accident_df.head()
```

```python
temp_df=us_accident_df[['Wind_Speed(mph)','Precipitation(in)']]
temp_df.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
sample_df=temp_df.sample(n=100000, random_state=42)
reg_imputer = IterativeImputer(random_state=42)
reg_imputer.fit(sample_df)
us_accident_df[['Wind_Speed(mph)', 'Precipitation(in)']] = reg_imputer.transform(
    us_accident_df[['Wind_Speed(mph)', 'Precipitation(in)']])

us_accident_df.head()
```

```python
us_accident_df[['Wind_Speed(mph)','Precipitation(in)']].head()
```

```python
us_accident_df.info()
```

```python
median_col=['Temperature(F)','Wind_Chill(F)', 'Humidity(%)', 'Visibility(mi)']
for col in median_col:
  us_accident_df[col]=us_accident_df[col].fillna(us_accident_df[col].median())

```

```python
us_accident_df[median_col].isnull().sum()
```

```python
us_accident_df['Pressure(in)']=us_accident_df['Pressure(in)'].fillna(us_accident_df['Pressure(in)'].mean())
```

```python
us_accident_df['Pressure(in)'].isnull().sum()
```

```python
us_accident_df['Wind_Direction']=us_accident_df['Wind_Direction'].fillna(us_accident_df['Wind_Direction'].mode()[0])
```

```python
us_accident_df['Wind_Direction'].isnull().sum()
```

```python
most_common = us_accident_df['Weather_Condition'].mode()[0]
us_accident_df['Weather_Condition'] = us_accident_df['Weather_Condition'].fillna(most_common)
```

```python
us_accident_df['Weather_Timestamp']=pd.to_datetime(us_accident_df['Weather_Timestamp'],errors='coerce')
us_accident_df['Weather_Timestamp']=us_accident_df['Weather_Timestamp'].ffill()
```

```python

us_accident_df['Weather_Timestamp'].isnull().sum()
```

```python
us_accident_df['Weather_Timestamp'].head()
```

```python
us_accident_df['Start_Time']=pd.to_datetime(us_accident_df['Start_Time'],format='mixed',errors='coerce')
us_accident_df['End_Time']=pd.to_datetime(us_accident_df['End_Time'],format='mixed',errors='coerce')
```

```python
#check if there are any duplicates
us_accident_df[us_accident_df.duplicated()].head()
```

```python
#ENCODING
# convert boolean columns to binary representation
bool_cols = [
    'Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit',
    'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming',
    'Traffic_Signal', 'Turning_Loop'
]
for col in bool_cols:
  us_accident_df[col]=us_accident_df[col].astype(int)
```

```python
# converting binary categorical columns to 0/1(binary representation)
binary_cols = ['Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']
for col in binary_cols:
  us_accident_df[col] = (us_accident_df[col]=='Day').astype(int)
```

```python
# applying Frequency Encoding to columns having less unique values

one_hot_cols = ['State', 'Timezone', 'Wind_Direction']
for col in one_hot_cols:
    freq = us_accident_df[col].value_counts()
    us_accident_df[col] = us_accident_df[col].map(freq)



```

```python
list(us_accident_df.columns)
```

```python
# label encoding for medium level unique values
from sklearn.preprocessing import LabelEncoder
Label_cols=['Weather_Condition', 'Airport_Code']
le=LabelEncoder()
for cols in Label_cols:
  us_accident_df[col]=le.fit_transform(us_accident_df[col].astype(str))
```

```python
#applying frequency encoding for high cardinality columns
freq_cols = ['City', 'County', 'Street', 'Zipcode','Airport_Code','Weather_Condition']
for col in freq_cols:
  freq = us_accident_df[col].value_counts()
  us_accident_df[col]=us_accident_df[col].map(freq)
```
