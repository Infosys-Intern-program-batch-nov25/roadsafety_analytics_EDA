```python
DATA_PATH = '/content/drive/MyDrive/Colab Notebooks/RoadSafety_Nov25/data/raw/US_Accidents_March23.csv'
import pandas as pd
df = pd.read_csv(DATA_PATH)
```

```python
# h0: Mean severity during clear weather equals mean severity during bad weather
# h1: Mean severity differs between clear and bad weather
from scipy.stats import ttest_ind

weather_df = df[['Severity', 'Weather_Condition']].dropna()

clear_weather = weather_df[weather_df['Weather_Condition'].str.contains('Clear', case=False)]['Severity']
bad_weather = weather_df[~weather_df['Weather_Condition'].str.contains('Clear', case=False)]['Severity']

t_stat, p_val = ttest_ind(clear_weather, bad_weather, equal_var=False)

alpha = 0.05
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.5f}")
print("Reject H0" if p_val < alpha else "Fail to reject H0")
```

```python
# h0: Visibility and severity are not correlated
# h1: Visibility and severity are correlated
from scipy.stats import spearmanr

vis_df = df[['Severity', 'Visibility(mi)']].dropna()

corr, p_val = spearmanr(vis_df['Severity'], vis_df['Visibility(mi)'])

alpha = 0.05
print(f"Spearman correlation: {corr:.3f}")
print(f"P-value: {p_val:.5f}")
print("Reject H0" if p_val < alpha else "Fail to reject H0")
```

```python
# h0: Severity level is independent of junction presence
# h1: Severity level is associated with junction presence
from scipy.stats import chi2_contingency

junction_df = df[['Severity', 'Junction']].dropna().copy()
junction_df['High_Severity'] = (junction_df['Severity'] >= 3).astype(int)

cont_table = pd.crosstab(junction_df['Junction'], junction_df['High_Severity'])
print(cont_table)

chi2, p_val, dof, expected = chi2_contingency(cont_table)

alpha = 0.05
print(f"Chi-square: {chi2:.3f}")
print(f"P-value: {p_val:.5f}")
print("Reject H0" if p_val < alpha else "Fail to reject H0")
```
