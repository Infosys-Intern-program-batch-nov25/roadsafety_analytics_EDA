### Load the US Accidents Dataset
- **Dataset Link:** [US Accident Dataset on Kaggle](https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents)

# US Accidents Dataset Column Definitions

| Column                | Description                                                                                 |
|-----------------------|---------------------------------------------------------------------------------------------|
| **ID**                | Unique identifier for each accident record                                                  |
| **Source**            | Source of the accident data                                                                 |
| **Severity**          | Integer representing the severity level of the accident                                    |
| **Start_Time**        | Timestamp when the accident started                                                        |
| **End_Time**          | Timestamp when the accident ended                                                          |
| **Start_Lat**         | Latitude coordinate where the accident started                                             |
| **Start_Lng**         | Longitude coordinate where the accident started                                            |
| **End_Lat**           | Latitude coordinate where the accident ended                                               |
| **End_Lng**           | Longitude coordinate where the accident ended                                              |
| **Distance(mi)**      | Distance in miles over which the accident event was recorded                               |
| **Description**       | Textual description of the accident                                                       |
| **Street**            | Name of the street where the accident occurred                                            |
| **City**              | City where the accident took place                                                        |
| **County**            | County where the accident occurred                                                        |
| **State**             | State of the accident location                                                            |
| **Zipcode**           | Postal code of the accident location                                                      |
| **Country**           | Country of the accident occurrence                                                        |
| **Timezone**          | Time zone of the accident location                                                        |
| **Airport_Code**      | Nearest airport code to the accident location                                             |
| **Weather_Timestamp** | Timestamp of the weather data related to the accident time                                |
| **Temperature(F)**    | Temperature in Fahrenheit at the time of the accident                                     |
| **Wind_Chill(F)**     | Wind chill temperature in Fahrenheit                                                      |
| **Humidity(%)**       | Humidity percentage during the accident                                                   |
| **Pressure(in)**      | Atmospheric pressure in inches                                                           |
| **Visibility(mi)**    | Visibility distance in miles at the time of the accident                                  |
| **Wind_Direction**    | Direction of the wind at the accident location                                            |
| **Wind_Speed(mph)**   | Wind speed in miles per hour during the accident                                         |
| **Precipitation(in)** | Amount of precipitation in inches                                                        |
| **Weather_Condition** | Description of the weather condition (e.g., clear, rainy, fog)                            |
| **Amenity**           | Boolean flag indicating if an amenity is present near the accident location               |
| **Bump**              | Boolean flag indicating if a bump is present near the accident location                   |
| **Crossing**          | Boolean flag indicating if a crossing is present near the accident location               |
| **Give_Way**          | Boolean flag indicating if a give-way sign is present                                    |
| **Junction**          | Boolean flag indicating if the accident happened at a junction                            |
| **No_Exit**           | Boolean flag indicating if there is no exit near the accident spot                        |
| **Railway**           | Boolean flag indicating if a railway crossing is near the accident                        |
| **Roundabout**        | Boolean flag indicating if a roundabout is near the accident                             |
| **Station**           | Boolean flag indicating if a station is near the accident                                |
| **Stop**              | Boolean flag indicating if a stop sign is near the accident                              |
| **Traffic_Calming**   | Boolean flag indicating if traffic calming measures exist near the accident              |
| **Traffic_Signal**    | Boolean flag indicating if traffic signals are present near the accident                  |
| **Turning_Loop**      | Boolean flag indicating if a turning loop is present near the accident                   |
| **Sunrise_Sunset**    | Indicates whether the accident occurred during sunrise or sunset periods                  |
| **Civil_Twilight**    | Indicates the civil twilight status at the time of the accident                          |
| **Nautical_Twilight** | Indicates the nautical twilight status at the time of the accident                       |
| **Astronomical_Twilight** | Indicates the astronomical twilight status at the time of the accident               |
