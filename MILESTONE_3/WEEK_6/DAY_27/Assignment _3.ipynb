```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/US_Accidents_March23.csv')

print("Dataset loaded successfully!")
```

```python
# First 5 rows
df.head()
# Dataset structure
df.info()
# Descriptive statistics
df.describe()
```

```python
weather_df = df[['Severity', 'Weather_Condition']].dropna()

clear = weather_df[weather_df['Weather_Condition'].str.contains('Clear', case=False)]
bad = weather_df[weather_df['Weather_Condition'].str.contains('Rain|Snow|Fog|Thunderstorm', case=False)]

t_stat, p_val = stats.ttest_ind(clear['Severity'], bad['Severity'], equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
if p_val < 0.05:
    print("Reject H₀ → Weather affects accident severity")
else:
    print("Fail to reject H₀")
```

```python
from scipy import stats
import pandas as pd

# Convert Start_Time safely
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')

# Drop rows where conversion failed
df = df.dropna(subset=['Start_Time'])

# Extract hour
df['Hour'] = df['Start_Time'].dt.hour

# Categorize time of day
def time_of_day(h):
    if 6 <= h < 12:
        return 'Morning'
    elif 12 <= h < 18:
        return 'Afternoon'
    elif 18 <= h < 24:
        return 'Evening'
    else:
        return 'Night'

df['Time_Category'] = df['Hour'].apply(time_of_day)

# Frequency counts
counts = df['Time_Category'].value_counts()

# Chi-square test
chi2, p = stats.chisquare(counts)

print("Chi-square:", chi2)
print("P-value:", p)
```

```python
if p < 0.05:
    print("Reject H₀ → Accident frequency depends on time of day")
else:
    print("Fail to reject H₀")
```

```python
road_df = df[['Severity', 'Junction']].dropna()

junction = road_df[road_df['Junction'] == True]['Severity']
non_junction = road_df[road_df['Junction'] == False]['Severity']

t_stat, p_val = stats.ttest_ind(junction, non_junction, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
if p_val < 0.05:
    print("Reject H₀ → Road features affect severity")
else:
    print("Fail to reject H₀")
```
