```python
#TASK
np.random.seed(42) # For reproducibility
n_samples = 100 # Number of samples

data = {
    'age': np.random.randint(18, 60, size=n_samples),
    'salary': np.random.randint(30000, 120000, size=n_samples),
    'department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing'], size=n_samples),
    'years_experience': np.round(np.random.normal(5, 2, size=n_samples), 1),
    'is_manager': np.random.choice([0, 1], size=n_samples)
}
df = pd.DataFrame(data)
```

```python
#Q1. View data structure
print(df.shape)             # rows, cols [web:3]
print(df.columns.tolist())  # column names [web:3]
print(df.dtypes)            # dtypes [web:3]
display(df.head())          # first rows [web:3]
```

```python
#Q2. Get DataFrame info and summary stats
df.info()  # concise schema summary [web:3]

# numeric summary
display(df.describe())

# categorical/binary summaries
display(df['department'].value_counts())  # frequency by department [web:3]
display(df['is_manager'].value_counts())  # manager vs non-manager counts [web:3]
```

```python
#Q3. Do simple NumPy operations
arr_salary = df['salary'].to_numpy()  # vector view [web:3]
salary_mean = np.mean(arr_salary)
salary_std  = np.std(arr_salary, ddof=1)

# z-score normalization
salary_z = (arr_salary - salary_mean) / salary_std

print("Mean salary:", salary_mean)
print("Sample std salary:", salary_std)
print("First 5 z-scores:", salary_z[:5])

# vectorized numeric feature: experience to salary ratio (per 1k)
ratio = df['years_experience'].to_numpy() / (arr_salary / 1000.0)
print("First 5 exp/salary(1k):", ratio[:5])
```

```python
#Q4. Filtering and indexing rows
# example filters
mask = (
    (df['department'].isin(['IT', 'Finance'])) &  # department filter [web:3]
    (df['salary'] > 80000) &                      # salary threshold [web:3]
    (df['years_experience'] >= 5)                 # experience threshold [web:3]
)
filtered = df.loc[mask, ['age','salary','department','years_experience','is_manager']]  # label indexing [web:3]
display(filtered.head())  # preview [web:3]

# positional indexing example
display(df.iloc[:5, :3])  # first 5 rows, first 3 columns [web:3]

# compound condition and sorted view
subset = df.loc[(df['is_manager']==1) & (df['salary']>=90000), ['department','salary','years_experience']]  # [web:3]
display(subset.sort_values('salary', ascending=False).head())  # [web:3]

```

```python
#Q5. Adding a column
df = df.copy()  # safe copy before mutation [web:3]

# derived features
df['salary_k'] = (df['salary'] / 1000.0).round(1)  # salary in thousands [web:3]
df['exp_per_year_pay_k'] = (df['years_experience'] / df['salary_k']).round(3)  # experience per 1k pay [web:3]
df['senior_flag'] = (df['years_experience'] >= 7).astype(int)  # binary flag [web:3]

display(df.head())  # confirm new columns [web:3]
```

```python
#Q6. Grouping and aggregation
# group by department with multiple aggregations
agg_dept = (
    df.groupby('department', observed=True)  # only show observed categories if categorical [web:3][web:2]
      .agg(
          avg_age=('age','mean'),
          avg_salary=('salary','mean'),
          median_exp=('years_experience','median'),
          managers=('is_manager','sum'),
          count=('age','count')
      )
      .round(2)
)
display(agg_dept)  # grouped stats per department [web:3][web:2]

# group by department and manager status
agg_dept_mgr = (
    df.groupby(['department','is_manager'], observed=True)
      .agg(
          avg_salary=('salary','mean'),
          n=('salary','count')
      )
      .round(2)
)
display(agg_dept_mgr)  # split-apply-combine result [web:5][web:3]

# pivot table for quick comparison
pivot = pd.pivot_table(
    df,
    values='salary',
    index='department',
    columns='is_manager',
    aggfunc='mean'
).round(2)  # mean salary by dept x manager [web:3]
display(pivot)
```
