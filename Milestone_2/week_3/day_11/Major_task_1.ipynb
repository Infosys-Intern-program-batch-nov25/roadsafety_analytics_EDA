```python
import pandas as pd
```

```python
dataset_path = '/content/drive/MyDrive/Colab Notebooks/RoadSafety_Nov25/data/raw/US_Accidents_March23.csv'
us_accidents_df = pd.read_csv(dataset_path)
```

```python
us_accidents_df.sample(10)
```

```python
#check which points in the col has missing vals, NaN gives False
us_accidents_df.isnull()
```

```python
#check total missing vals in each col
us_accidents_df.isnull().sum()
```

```python
#task given is to drop the 'End_Time' col
us_accidents_df = us_accidents_df.drop(columns=['End_Time'])
```

```python
us_accidents_df.sample(10)
```

```python
us_accidents_df['Start_Time'].dtype
```

```python
#convert 'Start_Time' to data time format
us_accidents_df['Start_Time'] = pd.to_datetime(
    us_accidents_df['Start_Time'],
    errors='coerce'
)
```

```python
us_accidents_df['Start_Time'].dtype
```

```python
#extract temporal info from 'Start_Time'
us_accidents_df['Year'] = us_accidents_df['Start_Time'].dt.year
us_accidents_df['Month'] = us_accidents_df['Start_Time'].dt.month
us_accidents_df['Day'] = us_accidents_df['Start_Time'].dt.day
us_accidents_df['Hour'] = us_accidents_df['Start_Time'].dt.hour
us_accidents_df['Weekday'] = us_accidents_df['Start_Time'].dt.day_name()
```

```python
us_accidents_df.head(10)
```

```python
#iterativeimputer for only int and float vals
numeric_cols = us_accidents_df.select_dtypes(include=['int64', 'float64']).columns
sample_df = us_accidents_df[numeric_cols].sample(n=100000, random_state=42)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

reg_imputer = IterativeImputer(random_state=42)
sample_imputed = reg_imputer.fit_transform(sample_df)
sample_imputed_df = pd.DataFrame(sample_imputed, columns=sample_df.columns)
sample_imputed_df.head()
```

```python
# Drop cols 'End_Lat', 'End_Lng', 'Wind_Chill(F)' and 'Precipitation(in)' as these cols contains lot of missing vals
cols_to_drop = ['End_Lat', 'End_Lng', 'Wind_Chill(F)', 'Precipitation(in)']
us_accidents_df.drop(columns=cols_to_drop, inplace=True)
```

```python
us_accidents_df.columns
```

```python
# For categorical encoding we need to identify all obj cols
categorical_cols = us_accidents_df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

onehot_cols = ['Source', 'State', 'Country', 'Weekday', 'Weather_Condition']
label_cols = ['City', 'County']

us_accidents_df = pd.get_dummies(us_accidents_df, columns=onehot_cols, drop_first=True)

for col in label_cols:
    le = LabelEncoder()
    us_accidents_df[col] = le.fit_transform(us_accidents_df[col].astype(str))

us_accidents_df.head()
```
